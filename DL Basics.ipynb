```python
'''Exercise 9 – Activation Functions 
After calculating the weighted sum (Z), we pass it through an Activation Function.
We'll use the same output (Z) from the previous exercise.

Create Sample Values'''

import numpy as np

# Output from previous layer (before activation)
z = np.array([-3, -1, 0, 1, 3])

print("Weighted Sum (Z):")
print(z)
```

```python
'''Sigmoid Function'''
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

sigmoid_output = sigmoid(z)

print("Sigmoid Output:")
print(sigmoid_output)
```

```python
'''ReLU Function'''
import numpy as np

def relu(x):
    return np.maximum(0, x)

relu_output = relu(z)

print("ReLU Output:")
print(relu_output)
```

```python
'''Tanh Function'''
import numpy as np

def tanh(x):
    return np.tanh(x)

tanh_output = tanh(z)

print("Tanh Output:")
print(tanh_output)
```

```python
'''Plot All Activation Functions'''
import numpy as np
import matplotlib.pyplot as plt

# Generate values
x = np.linspace(-5, 5, 200)

# Activation Functions
sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

# Plot
plt.figure(figsize=(8,5))

plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, relu, label="ReLU")
plt.plot(x, tanh, label="Tanh")

plt.title("Activation Functions")
plt.xlabel("Input (Z)")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

plt.show()
```

```python
'''Apply Activation to the Previous Exercise'''
import numpy as np

# Input (2 students, 3 features)
X = np.array([
    [2, 5, 8],
    [4, 6, 3]
])

# Weights
W = np.array([
    [0.2, 0.5],
    [0.4, 0.3],
    [0.6, 0.8]
])

# Bias
b = np.array([1, 2])

# Weighted Sum (Z)
Z = np.dot(X, W) + b

print("Before Activation (Z):")
print(Z)

# ReLU Activation
A = np.maximum(0, Z)
B = 1 / (1 + np.exp(-Z))
C = np.tanh(Z)

print("\nAfter ReLU Activation:")
print(A)
print("\nAfter Sigmoid Activation:")
print(B)
print("\nAfter Tanh Activation:")
print(C)
```
