# Batch Normalization vs Layer Normalization

## Objective

The objective of this experiment is to understand how Batch Normalization and Layer Normalization normalize input data and improve deep learning model training using TensorFlow.

```python
import numpy as np
import tensorflow as tf

# Make the output easier to read
np.set_printoptions(precision=3, suppress=True)
```

```python
# Generate random data
np.random.seed(42)

data = np.random.randint(10, 100, size=(5,4)).astype(np.float32)

print("Original Input:")
print(data)
```

```python
batch_norm = tf.keras.layers.BatchNormalization()

batch_output = batch_norm(data, training=True)

print("Batch Normalization Output:")
print(batch_output.numpy())
```

```python
layer_norm = tf.keras.layers.LayerNormalization()

layer_output = layer_norm(data)

print("Layer Normalization Output:")
print(layer_output.numpy())
```

```python
print("="*50)

print("Original Input")
print(data)

print("\nBatch Normalization")
print(batch_output.numpy())

print("\nLayer Normalization")
print(layer_output.numpy())
```

## Comparison

### Batch Normalization
Batch Normalization normalizes the input using the mean and variance of the entire batch of samples. It helps speed up training, stabilizes learning, and reduces internal covariate shift.

### Layer Normalization
Layer Normalization normalizes the features of each individual sample independently. It works well even when the batch size is small.

### Difference
Batch Normalization computes statistics across the batch, whereas Layer Normalization computes statistics across the features of each individual sample.
